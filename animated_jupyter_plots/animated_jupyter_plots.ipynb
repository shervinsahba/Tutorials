```python
from matplotlib import pyplot as plt
import numpy as np

# use this to clear the ipython cell output between plots
from IPython.display import clear_output
```

```python
def plot_this(x,y,t):
    clear_output(wait=True)
    fig,ax = plt.subplots(1,1,figsize=(6,2))
    ax.plot(x,y)
    ax.set_xlim(x[0],x[-1])
    plt.show()

x = np.linspace(0, 2*np.pi, num=100)
for t in range(20):
    plot_this(x,np.sin(x-t),t)
```
