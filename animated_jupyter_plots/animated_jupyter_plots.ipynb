Want an animated plot in Jupyter? Use `IPython.display.clear_output` during a plotting loop.

```python
from matplotlib import pyplot as plt
import numpy as np
from IPython.display import clear_output
from time import sleep
```

```python
x = np.linspace(0, 2*np.pi, num=100)

for t in range(30):
    clear_output(wait=True)  # clears IPython during loop
    fig,ax = plt.subplots(1,1,figsize=(6,2))
    ax.plot(x,np.cos(x-t))
    ax.set_xlim(x[0],x[-1])
    ax.set_ylim(-1,1)
    plt.show()
    sleep(0.02)  # not necessary. use it to slow down the loop.
```

This probably isn't optimized, since it clearing the figure on each iteration. But, hey, it works.
